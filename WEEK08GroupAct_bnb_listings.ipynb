```python
import numpy as np
import pandas as pd

#modules for statistics
import math
import statistics
from scipy import stats

#belize listings csv file into dataframe
belize_df=pd.read_csv('belize_airbnb_listings.csv')
```

# Arithmetic Mean
The Arithmetic mean is actually the sum of all observations divided by number of observations.


```python
A_mean = belize_df['price'].mean()
A_mean
```

# Geometric Mean
The geometric mean is a mean, which specifies the central tendency of a set of numbers by using the multiply of their values.

```python
G_mean = stats.gmean(belize_df['price'])
G_mean
```

# Harmonic Mean
The harmonic mean is a type of numerical average. It is calculated by dividing the number of observations by the reciprocal of each number in the series. 

Mathematical formula for arithmetic, geometric and harmonic mean


```python
H_mean = stats.hmean(belize_df['price'])
H_mean
```
